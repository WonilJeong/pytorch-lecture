### 유용한 단축키

**[실행 관련 단축키]**

- **Ctrl + Enter** = 해당 셀을 실행하고 커서를 해당 셀에 두는 경우 (결과 값만 보고자 할 때)

- **Shift + Enter** = 해당 셀을 실행하고 커서를 다음 셀로 넘기는 경우 (여러가지 값을 빠르게 출력할 때)

- **Alt + Enter** = 해당 셀을 실행하고 셀을 삽입한 후 커서를 삽입한 셀로 넘기는 경우 (다음 작업 공간이 없을 때)


**[셀 삽입/삭제 관련 단축키]**

**(Ctrl + M +)**

- **A** = 코드 셀 위에 삽입

- **B** = 코드 셀 아래 삽입

- **D** = 셀 지우기

- **Y** = 코드 셀로 변경

- **M** = 마크다운 셀로 변경

- **Z** = 실행 취소

- **H** = 단축키 보기

### 제어문

```python
x = 2

if x > 0:
    print('값이 0보다 큽니다.')
```

```python
# (문제) if elif else 를 이용해 x에 따라 '값이 0보다 큽니다' '값이 0입니다.' '값이 0보다 작습니다' 출력하기
x = -1

if x > 0:
    print('값이 0보다 큽니다')
elif x == 0:
    print('값이 0입니다.')
else:
    print('값이 0보다 작습니다')
```

### 함수

```python
def f(x):
    y = 2 * x + 1;
    return y

print(f(1))
print(f(3))
```

```python
# (문제) 더하기 함수 adder()  만들기
def adder(a, b):
    return a+b

print(adder(2,3))
print(adder(4,8))
```

```python
# (문제) 제곱근 함수 sqrt()  만들기
def sqrt(x):
    res = x**(1 / 2)
    return res

print(sqrt(2))
print(sqrt(4))
print(sqrt(9))
```

### 패키지 사용

```python
import pandas

raw_data = {'col0': [1, 2, 3, 4],
            'col1': [10, 20, 30, 40],
            'col2': [100, 200, 300, 400]}

data = pandas.DataFrame(raw_data)
print(data)
```

```python
# (문제) pandas 를 pd로 import 하기
import pandas as pd

raw_data = {'col0': [1, 2, 3, 4],
            'col1': [10, 20, 30, 40],
            'col2': [100, 200, 300, 400]}

data = pd.DataFrame(raw_data)
print(data)
```

```python
# (문제) pandas.DataFrame을 df로 import 하기
from pandas import DataFrame as df

raw_data = {'col0': [1, 2, 3, 4],
            'col1': [10, 20, 30, 40],
            'col2': [100, 200, 300, 400]}

data = df(raw_data)
print(data)
```

Numpy 연산

```python
x = np.array([1,2,3])
y = np.array([4,5,6])
```

```python
result = np.add(x, y)
print(result)

result = x + y
print(result)
```

```python
# result = np.subtract(x, y)와 동일.
result = x - y
print(result)
```

```python
# result = np.multiply(x, y)와 동일.
result = x * y
print(result)
```

```python
# result = np.divide(result, x)와 동일.
result = result / x
print(result)
```

```python
result = np.dot(x, y)
print(result)
```

```python
mat1 = np.array([[1,2],[3,4]])
mat2 = np.array([[5,6],[7,8]])
mat3 = np.dot(mat1, mat2)
print(mat3)
```

#### Numpy 연습문제

```python
# (문제) 3*4크기의 0으로 이루어진 배열 생성 후, [1,2] 위치의 값을 1으로 변경
mat = np.zeros((3,4))
mat[1,2] = 1
print(mat)
```

```python
# (문제) 0~49의 배열 생성 후 5*10의 matrix로 reshape
mat = np.array(np.arange(50)).reshape((5,10))
print(mat)
```

```python
# 배열 [2, 4, 8, 16] 과 [3, 6, 9, 12] 생성 후 두 벡터의 곱과 내적 구하기
a = np.array([2, 4, 8, 16])
b = np.array([3, 6, 9, 12])

multiplied = np.multiply(a, b)
dot_product = np.dot(a, b)

print(multiplied)
print(dot_product)
```
